## Damage assesment: SSM, TAUW, WSS

```python
flood_damage_per_sqm = 271  
# Define a mapping function
def map_water_depth(value):
    mapping = {
        0: 0.0, 
        1: 0.10, #10cm
        2: 0.15, #15cm
        3: 0.20, #20cm
        4: 0.30, #30cm
        5: 0.50  #50cm
    }
    return mapping.get(value, 0)  


for id in residential_rows['address_id'][1:4]:
    print(f'id:      {id}')
    # Get the address_id indices
    address_pixels = np.argwhere(ds['address_id'].values == id)

    # Get the buffer indices (2-pixel buffer since 0.5m resolution)
    buffer_indices = []
    
    for pixel in address_pixels:
        x, y = pixel
        # Create a buffer around each pixel (2-pixel buffer since 0.5m resolution)
        for i in range(x - 2, x + 3):
            for j in range(y - 2, y + 3):
                # Use ds.sizes instead of ds.dims
                if (i >= 0 and i < ds.sizes['x']) and (j >= 0 and j < ds.sizes['y']):
                    buffer_indices.append((i, j))

    buffer_indices = np.unique(buffer_indices, axis=0)  # Remove duplicates

    # Calculate building area and perimeter
    surface_area = residential_rows.loc[residential_rows['address_id'] == id, 'building_area'].values[0]
    perimeter    = residential_rows.loc[residential_rows['address_id'] == id, 'building_perimeter'].values[0]
    
    print(f'surface_area: {surface_area}')
    print(f'perimeter:    {perimeter}')


    # Apply the function to the array using np.vectorize
    water_depth = np.vectorize(map_water_depth)(ds['70mm'].values[[tuple(idx) for idx in buffer_indices]])    

    land_height = ds['AHN'].values[[tuple(idx) for idx in buffer_indices]]

    # Calculate the average water depth in the buffer
    avg_water_depth_buffer = np.mean(water_depth)
    print(f'avg_water_depth_buffer: {avg_water_depth_buffer}')
    
    # Check if the building is considered flooded
    if avg_water_depth_buffer > 0.15:
        # The whole building is considered flooded if average water depth in the buffer > 15 cm
        residential_building = WSS.Building(surface_area, perimeter, building_value_per_m2=flood_damage_per_sqm)
        total_damage = surface_area * flood_damage_per_sqm
        print(f"Flooded, Estimated Damage: {total_damage} euros.\n")
    
    else:
        # Interpolate water depth inside the building if not fully flooded
        max_water_depth = np.max(water_depth)
        min_water_depth = np.min(water_depth)

        # Check minimum water depth against the facade
        water_volume_against_facade = max_water_depth * surface_area

        if water_volume_against_facade / surface_area >= 0.02:
            # Interpolate water depth inside the building
            interpolated_water_depth = (max_water_depth + min_water_depth) / 2
            # Compute potential damage based on interpolated water depth
            building_damage_factor = interpolated_water_depth / 0.15  # Scaling factor based on water depth
            total_damage = building_damage_factor * surface_area * flood_damage_per_sqm
            print(f"Partial Flooding, Estimated Damage: {total_damage} euros\n")
        else:
            print(f"No significant flood damage.\n")
```

```python
def calculate_damage(H_out: float, H_threshold: float = 0.15, S_max: float = 250) -> float:
    """
    Calculate the direct structural damage to a residential building based on water depth.

    Args:
    - H_out (float):                    Water depth outside the building in meters.
    - H_threshold (float, optional):    Threshold height to prevent water ingress in meters. Default is 0.15 meters.
    - S_max (float, optional):          Maximum direct damage per square meter in euros. Default is 250 €/m².

    Returns:
    - S_damage (float): The calculated direct damage per square meter in euros.
    """
    # Step 1: Calculate the water depth inside the building (H_in)
    if H_out <= H_threshold:
        H_in = 0
    else:
        H_in = H_out - H_threshold

    # Step 2: Define the damage factor (alpha) based on H_in
    if H_in == 0:
        alpha = 0
    elif H_in <= 0.30:
        alpha = H_in / 0.30
    else:
        alpha = 1  # Maximum damage

    # Step 3: Calculate the direct damage per square meter (S_damage)
    S_damage = alpha * S_max

    return S_damage


# Example calculation
H_out_example = 0.35  # Example water depth outside the building in meters
S_damage_example = calculate_damage(H_out_example)

print(f"Calculated damage per square meter: €{S_damage_example:.2f}/m²")
```
